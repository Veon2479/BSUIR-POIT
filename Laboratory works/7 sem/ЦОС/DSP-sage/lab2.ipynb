```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def getHarmony(N, M, fi):
    list = np.empty(N)
    for i in range(N):
        list[i] = math.sin((2*math.pi*i)/N + fi)
    return list
```

```python
N = 2^10
x = range(N)
M = 2 * N

K = N / 4 * 3
fi = math.pi/32
data = getHarmony(N, M, 0)

plt.plot(x, data)
```
