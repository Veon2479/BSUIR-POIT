```python
import random as rn
import numpy as np
import matplotlib.pyplot as plt

N = 1000000
```

```python
def Mean(data):
    mean = 0.0
    for x in data:
        mean = mean + x
    return mean/len(data)
```

```python
def Dispersion(data):
    dispersion = 0.0
    mean = Mean(data)
    for x in data:
        dispersion = dispersion + (x - mean)**2
    dispersion = dispersion / len(data)
    return dispersion
```

```python
def Deviation(data):
    return Dispersion(data)**0.5
```

```python
def info(data):
    print(f'Mean is {Mean(data)}')
    print(f'Dispersion is {Dispersion(data)}')
    print(f'Deviation is {Deviation(data)}')
    plt.hist(data, bins=30)
    plt.show() 
```

```python
r = []
for i in range(N):
    r.append(rn.random())
r = np.array(r)
```

### Равномерное

```python
a1 = 2
b1 = 10
A1 = np.array((b1-a1)*r) + a1
info(A1)
```

### Нормальное

```python
m2 = 6
d2 = 10
r2 = np.array_split(r, len(r)//6)
A2 = []
for a in r2:
    sum = 0
    for i in a:
        sum += i
    A2.append(m2 + d2*(2**0.5)*(sum - 3))
info(A2)
```

### Экспоненциальное

```python
l3 = 4
A3 = - np.log(np.array(r))/l3
info(A3)
```

### Гамма

```python
n4 = 3
u4 = 6
rr4 = np.array_split(r, len(r)//n4)
A4 = []
for a in rr4:
    tmp = 1
    for i in a:
        tmp *= i
    A4.append(-np.log(tmp)/u4)
info(A4)
```

### Треугольное

```python
a5 = 3
b5 = 7
r5 = np.array_split(r, len(r)//2)
A5 = []
for a in r5:
    A5.append(a5 + (b5 - a5)*np.max(a))
info(A5)
```

### Симпсона

```python
a6 = 4
b6 = 6
R6 = np.array_split(r, 2)
R61 = R6[0]
R62 = R6[1]
A61 = np.array((b6/2-a6/2)*R61) + a6/2
A62 = np.array((b6/2-a6/2)*R62) + a6/2
A6 = []
for i in range(len(A61)):
    A6.append(A61[i]+A62[i])
info(A6)
```
