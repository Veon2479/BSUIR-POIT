```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a = 19
# R0 = 37
# m = 127

# a = 125566
# m = 276128
# R0 = 1

a = 125566
m = 276128
R0 = 1

# a = 3
# m = 12
# R0 = 9

R = R0

def random():
    global R, a, m
    R = (a * R) % m;
    return R
```

```python
N = 1000000
```

```python
mean = 0.0
for i in range(N):
    mean += random()
mean /= float(m)
mean /= float(N)

print(f'Mean is {mean}')
```

```python
deviation = 0.0
for i in range(N):
    deviation += (mean - float(random())/float(m))**2
deviation /= N

dispersion = deviation**0.5

print(f'Deviation is {deviation}')
print(f'Dispersion is {dispersion}')
```

```python
data = []
for i in range(N):
    data.append(random()/float(m))
plt.hist(data, bins=30)
plt.show()
```

```python
count = 0
for i in range(N):
    t = (random()/float(m))**2 + (random()/float(m))**2
    if t < 1:
        count-=-1
print(f'{count/N} ~ {3.1415/4}')
```

```python
# R = R0
# for i in range(200000):
#     random()

xv = random()
R = R0

i1 = 0
t = -1
while (t != xv):
    t = random()
    i1-=-1
i2 = i1
t=-1
while (t != xv):
    t = random()
    i2-=-1
P = i2 - i1
print(f'Period is {P}')
```

```python
R = R0
data = []
for i in range(P+1):
    data.append(random())

while data[-1] != data[-1 - P]:
    data.append(random())

print(f'Aperiodic is {len(data)-1}')
```
